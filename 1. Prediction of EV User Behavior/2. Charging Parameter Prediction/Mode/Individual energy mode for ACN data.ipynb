```python
def convert_to_interval(datetime_str):
    # Extract time string from the datetime string
    time_str = datetime_str.split()[1]

    # Convert time string into hours, minutes, and seconds
    hours, minutes, seconds = map(int, time_str.split(':'))

    # Calculate total minutes
    total_minutes = hours * 60 + minutes

    # Find the nearest lower multiple of 15 for the minutes
    interval_minutes = (total_minutes // 15) * 15

    # Convert back to hour and minute
    interval_hour, interval_minute = divmod(interval_minutes, 60)

    # Return as a formatted string
    return "{:02d}:{:02d}".format(interval_hour, interval_minute)

# Apply the conversion to the data
sorted_caltech_training[:, 0] = np.vectorize(convert_to_interval)(sorted_caltech_training[:, 0])
sorted_caltech_testing[:, 0] = np.vectorize(convert_to_interval)(sorted_caltech_testing[:, 0])

sorted_JPL_training[:, 0] = np.vectorize(convert_to_interval)(sorted_JPL_training[:, 0])
sorted_JPL_testing[:, 0] = np.vectorize(convert_to_interval)(sorted_JPL_testing[:, 0])
```

```python
#Mode calculation with repetition

from collections import Counter
def compute_mode_per_user_per_interval(data):
    """Compute the mode per user per time interval, filling gaps with nearest mode."""
    user_groups = {}
    mode_per_user_per_interval = {}

    # Group by user
    for row in data:
        if row[2] not in user_groups:
            user_groups[row[2]] = []
        user_groups[row[2]].append(row)

    # Calculate modes per user
    for user, entries in user_groups.items():
        time_intervals = sorted(list(set([entry[0] for entry in entries])))
        modes = {}
        for interval in time_intervals:
            session_values = [entry[1] for entry in entries if entry[0] == interval]
            counts = Counter(session_values)
            highest_freq = max(counts.values())
            common_vals = [key for key, val in counts.items() if val == highest_freq]

            # If there's a single mode, use it; else, average them
            modes[interval] = sum(common_vals) / len(common_vals)

        # Fill gaps in modes using nearest mode
        all_intervals = sorted(list(set(data[:, 0])))
        for interval in all_intervals:
            if interval not in modes:
                # Search backward for mode
                prev_mode = next((modes[prev_int] for prev_int in reversed(all_intervals[:all_intervals.index(interval)]) if prev_int in modes), None)

                # If no previous mode found, search forward
                if prev_mode is None:
                    prev_mode = next((modes[next_int] for next_int in all_intervals[all_intervals.index(interval)+1:] if next_int in modes), None)

                if prev_mode is not None:
                    modes[interval] = prev_mode

        mode_per_user_per_interval[user] = modes

    return mode_per_user_per_interval

# Update mode_per_user_per_interval using the new function
mode_per_user_per_interval_caltech_repeat = compute_mode_per_user_per_interval(sorted_data_caltech_training)
mode_per_user_per_interval_JPL_repeat = compute_mode_per_user_per_interval(sorted_data_JPL_training)
```

```python
#Mean calculation with repetition
from collections import defaultdict
import numpy as np

def compute_mean_per_user_per_interval(data):
    """Compute the mean per user per time interval, with adjustments for gaps."""
    user_groups = defaultdict(list)
    mean_per_user_per_interval = {}

    # Group by user
    for row in data:
        user_groups[row[2]].append(row)

    # Calculate means per user
    for user, entries in user_groups.items():
        time_intervals = sorted(list(set([entry[0] for entry in entries])))
        means = {}
        for interval in time_intervals:
            session_values = [entry[1] for entry in entries if entry[0] == interval]
            means[interval] = np.mean(session_values)

        # Fill gaps in means using nearest non-empty interval if applicable
        all_intervals = sorted(list(set(data[:, 0])))
        for interval in all_intervals:
            if interval not in means:
                # Search backward for mean
                prev_mean = next((means[prev_int] for prev_int in reversed(all_intervals[:all_intervals.index(interval)]) if prev_int in means), None)

                # If no previous mean found, search forward
                if prev_mean is None:
                    prev_mean = next((means[next_int] for next_int in all_intervals[all_intervals.index(interval)+1:] if next_int in means), None)

                if prev_mean is not None:
                    means[interval] = prev_mean

        mean_per_user_per_interval[user] = means

    return mean_per_user_per_interval

# Update mean_per_user_per_interval using the new function for both datasets
mean_per_user_per_interval_caltech_repeat = compute_mean_per_user_per_interval(sorted_data_caltech_training)
mean_per_user_per_interval_JPL_repeat = compute_mean_per_user_per_interval(sorted_data_JPL_training)
```

```python
#Mode calculation with linear interpolation

def compute_mode_per_user_per_interval(data):
    """Compute the mode per user per time interval with linear interpolation."""
    user_groups = {}
    mode_per_user_per_interval = {}

    # Group by user
    for row in data:
        if row[2] not in user_groups:
            user_groups[row[2]] = []
        user_groups[row[2]].append(row)

    # Calculate modes per user
    for user, entries in user_groups.items():
        time_intervals = sorted(list(set([entry[0] for entry in entries])))
        modes = {}
        for interval in time_intervals:
            session_values = [entry[1] for entry in entries if entry[0] == interval]
            counts = Counter(session_values)
            highest_freq = max(counts.values())
            common_vals = [key for key, val in counts.items() if val == highest_freq]

            # If there's a single mode, use it; else, average them
            modes[interval] = sum(common_vals) / len(common_vals)

        # Linearly interpolate for missing modes
        all_intervals = sorted(list(set(data[:, 0])))
        for i, interval in enumerate(all_intervals):
            if interval not in modes:
                # Find previous and next known modes
                prev_mode = next((modes[prev_int] for prev_int in reversed(all_intervals[:i]) if prev_int in modes), None)
                next_mode = next((modes[next_int] for next_int in all_intervals[i+1:] if next_int in modes), None)

                # If both previous and next modes exist, interpolate
                if prev_mode is not None and next_mode is not None:
                    gap_size = all_intervals[i+1:].index(next((next_int for next_int in all_intervals[i+1:] if next_int in modes))) + 1
                    increment = (next_mode - prev_mode) / (gap_size + 1)
                    modes[interval] = prev_mode + increment

                # If no next mode exists, keep the mode same as the previous mode
                elif prev_mode is not None:
                    modes[interval] = prev_mode

                # If no previous mode exists, keep the mode same as the next mode
                elif next_mode is not None:
                    modes[interval] = next_mode

        mode_per_user_per_interval[user] = modes

    return mode_per_user_per_interval

# Update mode_per_user_per_interval using the new function
mode_per_user_per_interval_caltech_interpolate = compute_mode_per_user_per_interval(sorted_data_caltech_training)
mode_per_user_per_interval_JPL_interpolate = compute_mode_per_user_per_interval(sorted_data_JPL_training)
```

```python
#Mean calculation with linear interpolation

def compute_mean_per_user_per_interval_with_interpolation(data):
    """Compute the mean per user per time interval with linear interpolation."""
    user_groups = {}
    mean_per_user_per_interval = {}

    # Group by user
    for row in data:
        if row[2] not in user_groups:
            user_groups[row[2]] = []
        user_groups[row[2]].append(row)

    # Calculate means per user
    for user, entries in user_groups.items():
        time_intervals = sorted(list(set([entry[0] for entry in entries])))
        means = {}
        for interval in time_intervals:
            session_values = [entry[1] for entry in entries if entry[0] == interval]
            mean_value = sum(session_values) / len(session_values) if session_values else None
            if mean_value is not None:
                means[interval] = mean_value

        # Linearly interpolate for missing means
        all_intervals = sorted(list(set([entry[0] for entry in data])))
        for i, interval in enumerate(all_intervals):
            if interval not in means:
                # Find previous and next known means
                prev_mean = next((means[prev_int] for prev_int in reversed(all_intervals[:i]) if prev_int in means), None)
                next_mean = next((means[next_int] for next_int in all_intervals[i+1:] if next_int in means), None)

                # If both previous and next means exist, interpolate
                if prev_mean is not None and next_mean is not None:
                    gap_size = all_intervals[i+1:].index(next((next_int for next_int in all_intervals[i+1:] if next_int in means))) + 1
                    increment = (next_mean - prev_mean) / (gap_size + 1)
                    means[interval] = prev_mean + increment

                # If no next mean exists, keep the mean the same as the previous mean
                elif prev_mean is not None:
                    means[interval] = prev_mean

                # If no previous mean exists, keep the mean the same as the next mean
                elif next_mean is not None:
                    means[interval] = next_mean

        mean_per_user_per_interval[user] = means

    return mean_per_user_per_interval

# Update mean_per_user_per_interval with interpolation using the new function for both datasets
mean_per_user_per_interval_caltech_interpolate = compute_mean_per_user_per_interval_with_interpolation(sorted_data_caltech_training)
mean_per_user_per_interval_JPL_interpolate = compute_mean_per_user_per_interval_with_interpolation(sorted_data_JPL_training)
```
